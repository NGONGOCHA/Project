```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#Hàm trích xuất Tiêu đề sản phẩm
def get_title(soup):

    try:
        # Đối tượng thẻ bên ngoài
        title = soup.find("span", attrs={"id":'productTitle'})

        # Đối tượng text bên trong
        title_value = title.text

        # Tiêu đề dưới dạng giá trị chuỗi
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string

# Chức năng trích xuất giá sản phẩm
def get_price(soup):

    try:
        price = soup.find("span", attrs={'class':'a-price'}).string.strip()

    except AttributeError:

        try:
            # Nếu có một số giá thỏa thuận
            price = soup.find("span", attrs={'class':'a-offscreen'}).string.strip()

        except:
            price = ""

    return price

# Chức năng trích xuất Xếp hạng sản phẩm
def get_rating(soup):

    try:
        rating = soup.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5'}).string.strip()

    except AttributeError:
        try:
            rating = soup.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            rating = ""

    return rating

# Chức năng trích xuất số lượng đánh giá của người dùng
def get_review_count(soup):
    try:
        review_count = soup.find("span", attrs={'id':'acrCustomerReviewText'}).string.strip()

    except AttributeError:
        review_count = ""

    return review_count

# Chức năng trích xuất Trạng thái sẵn có
def get_availability(soup):
    try:
        available = soup.find("div", attrs={'id':'availability'})
        available = available.find("span").string.strip()

    except AttributeError:
        available = "Not Available"

    return available
```

```python
if __name__ == '__main__':

    d = {"links_list":[], "title":[], "price":[], "rating":[], "reviews":[],"availability":[]}

    # thêm tác nhân người dùng của bạn
    HEADERS = ({'User-Agent':'', 'Accept-Language': 'en-US, en;q=0.5'})

    #URL trang web
    URL = "https://www.amazon.com/s?k=playstation+5&crid=3DUO539EKKRRZ&qid=1701171914&sprefix=playstation+%2Caps%2C446&ref=sr_pg_1"

    #Yêu cầu HTTP
    webpage = requests.get(URL, headers=HEADERS)

    # Đối tượng Soup chứa tất cả dữ liệu
    soup = BeautifulSoup(webpage.content, "html.parser")

    links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})

      # Vòng lặp trích xuất liên kết từ Đối tượng thẻ
    for link in links:
            d['links_list'].append("https://www.amazon.com" + link.get('href'))

    # Vòng lặp trích xuất chi tiết sản phẩm từ mỗi liên kết
    for link in d['links_list']:
        new_webpage = requests.get(link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Lệnh gọi hàm hiển thị tất cả thông tin sản phẩm cần thiết
        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['reviews'].append(get_review_count(new_soup))
        d['availability'].append(get_availability(new_soup))


    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('', np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data1.csv", header=True, index=False)

amazon_df
```
